```python
from fastai import *
from fastai.vision import *
```

## Get the data

```python
path = Path('../data/carvana/')
path.ls()
```

```python
filenames = get_image_files(path/'train')
filenames[:3]
```

```python
masks = get_image_files(path/'train_masks')
masks[:3]
```

```python
img = open_image(filenames[0])
img.show(figsize=(5,5))
```

function to return the path to the mask for an image in the training set

```python
def get_mask(y): return Path(str(y.parent) + '_masks') / (y.name[:-4] + '_mask.gif')
```

```python
mask = open_mask(get_mask(filenames[0]))
mask.show(figsize=(5,5), alpha=1)
```

```python
src_size = np.array(mask.shape[1:])
src_size, mask.data
```

### create a databunch

```python
bs, size = 8, src_size // 2 
size[1] += 1 # adding one makes all sides even
bs, size
```

```python
codes = np.array(['background', 'car'])
```

create custom classes to allow for binary segmentation

```python
class SegLabelListCustom(SegmentationLabelList):
    def open(self, fn): return open_mask(fn, div=True)
    
class SegItemListCustom(ImageItemList):
    _label_cls = SegLabelListCustom
```

```python
src = (SegItemListCustom.from_folder(path/'train')
       .random_split_by_pct()
       .label_from_func(get_mask, classes=codes))
```

```python
data = (src.transform(get_transforms(), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(imagenet_stats))
```

```python
data.show_batch(rows=2, figsize=(10, 7), alpha=1)
```

## Model

```python
learn = unet_learner(data, models.resnet34, metrics=dice, wd=1e-2)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr = 1e-2
```

```python
learn.fit_one_cycle(5, slice(lr))
```
